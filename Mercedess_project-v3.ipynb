### 9.1. XGBoost Hyperparameter Tuning:

**Hyperparameters**:

- **n_estimators** [default=100] - the number of trees in the ensemble.

- **learning_rate** [default=0.3] - shrinks the three weights in each round of boosting. (Decreasing prevents overfitting)

- **min_child_weight** [default=1] - the minimum sum of weights required for the node to split. (Increasing prevents overfitting)
Too high values can lead to under-fitting hence, it should be tuned using CV.

- **max_depth** [default=6] - the depth of the three. (Decreasing prevents overfitting)
Should be tuned using CV.
Typical values: 3-10

- **gamma** [default=0] - a node is split only when the resulting split gives a positive reduction in the loss function. Gamma specifies the minimum loss reduction required to make a split. (Increase prevents overfitting. Low values, usually lower than 10, are standard)

- **max_delta_step** [default=0] - only recommended for extremly imbalanced datasets. In maximum delta step we allow each trees weight estimation to be. (Increasing prevents overfitting)

- **subsample** [default=1] - limits the persentage of training rows for each boosting round. Denotes the fraction of observations to be randomly samples for each tree.(Decreasing prevents overfitting)
Typical values: 0.5-1

- **colsample_bytree** [default=1] - limits the persentage of training columns for each boosting round. (Decreasing prevents overfitting)
Typical values: 0.5-1

- **colsample_bylevel** [default=1] - limits the persentage of  columns for each depth level of the three. Denotes the subsample ratio of columns for each split, in each level. (Decreasing prevents overfitting)

- **reg_lambda** [default=1] - L2 regularization term on weights (analogous to Ridge regression). (Increasing prevents overfitting)

- **alpha** [default=0] - L1 regularization term on weight (analogous to Lasso regression). (Increasing prevents overfitting)

- **scale_pos_weight** [default=1] - used for imbalaced datasets. A value greater than 0 should be used in case of high class imbalance as it helps in faster convergence.

## Stratification on 10 stratas:

```python
from sklearn.model_selection import KFold
from sklearn.metrics import mean_squared_error, r2_score, accuracy_score

N=10

train_list = []
test_list = []

X_train_fold = []
X_test_fold = []
y_train_fold = []
y_test_fold = []


# stratifiedKfold by department
kfold = KFold(n_splits=N, random_state=42,shuffle=True)
X = df[total_features]
y = df['y']

for i, (train_index,test_index) in enumerate(kfold.split(X)):
    X_train = X.iloc[train_index]
    X_test = X.iloc[test_index]
    y_train = y.iloc[train_index]
    y_test = y.iloc[test_index]

    X_train_fold.append((X_train))
    X_test_fold.append((X_test))
    y_train_fold.append((y_train))
    y_test_fold.append((y_test))
```

# Using each strata to get predictions for test.csv data:

```python
predictions=[]
rmses=[]
r2_scores = []
y_train = []
X_train = []
y_test=[]
X_test =[]

for  i in range(N):
    pre_fin = []
    
    
    model = XGBRegressor(alpha= 0.2514743295310575, gamma= 1, learning_rate= 0.029234073208504588, max_depth= 3, min_child_weight= 1,
                         n_estimators= 192, reg_lambda= 2, scale_pos_weight= 2.3011397801271274, random_state=42)

    
    y_train = y_train_fold[i]
    X_train = X_train_fold[i]
    
    y_test = y_test_fold[i]
    X_test = X_test_fold[i]
    
    X_train = transformer_pipe_skb.fit_transform(X_train,y_train)
    X_test = transformer_pipe_skb.transform(X_test)
    
   
    # fit the model
    model.fit(X_train,y_train,
             early_stopping_rounds=100, 
             eval_set=[(X_test, y_test)],
             verbose=0)
    
    # evaluate on test data
    pre = model.predict(X_test)
    
    rmse = mean_squared_error(y_test,pre,squared=False)
    rmses.append(rmse)
    
    acc = r2_score(y_test, pre)
    r2_scores.append(acc)
    
    print(f'Fold {i+1}')
    print('RMSE: {:.2f}'.format(rmse))
    print('R2: {:.2f}'.format(acc))
    
    print()
    # two votes if scored well 
    pre_fin = model.predict(test_df)
    predictions.append(pre_fin)
print()
print(f'Average RMSE: {np.mean(rmses)}')
print(f'Average R2: {np.mean(r2_scores)}')
```
